There is a horizontal row of  cubes.

```python
for _ in range(int(input())):
    n = int(input())
    cubes = list(map(int, input().split()))
    top = max(cubes) + 1
    left = 0
    right = n - 1
    while left <= right:
        if cubes[left] <= top and (cubes[left] >= cubes[right] or cubes[right] > top):
            top = cubes[left]
            left += 1
        elif cubes[right] <= top:
            top = cubes[right]
            right -= 1
        else:
            break
    print("Yes" if left > right else "No")
```

set_mutations

```python
n = int(input())
A = set()
for _ in range(n):
    A.add(int(input()))

N = int(input())
for _ in range(N):
    line = input().split()
    op = line[0]
    m = int(line[1])
    other = set()
    for _ in range(m):
        other.add(int(input()))

    if op == "update":
        A.update(other)
    elif op == "intersection_update":
        A.intersection_update(other)
    elif op == "difference_update":
        A.difference_update(other)
    elif op == "symmetric_difference_update":
        A.symmetric_difference_update(other)

total = 0
for x in A:
    total += x

print(total)
```

finding precentage

```python
n = int(input())
student_marks = {}

for _ in range(n):
    data = input().split()
    name = data[0]
    marks = []
    for mark in data[1:]:
        marks.append(float(mark))  # تحويل كل درجة لعدد عشري
    student_marks[name] = marks

query_name = input()

marks_list = student_marks[query_name]
average = sum(marks_list) / len(marks_list)

print("{:.2f}".format(average))
```

swaping

```python
def swap_case(s):
    res = ""
    for c in s:
        if c.islower():
            res += c.upper()
        elif c.isupper():
            res += c.lower()
        else:
            res += c
    return res

s = input()
print(swap_case(s))
```

```python
if __name__ == '__main__':
    # N is the number of commands
    try:
        N = int(input().strip())
    except EOFError:
        N = 0

    my_list = []

    for _ in range(N):
        # .split() handles multiple spaces and breaks the line into parts
        args = input().strip().split()
        command = args[0]

        if command == "insert":
            my_list.insert(int(args[1]), int(args[2]))
        elif command == "print":
            print(my_list)
        elif command == "remove":
            my_list.remove(int(args[1]))
        elif command == "append":
            my_list.append(int(args[1]))
        elif command == "sort":
            my_list.sort()
        elif command == "pop":
            my_list.pop()
        elif command == "reverse":
            my_list.reverse()
```
